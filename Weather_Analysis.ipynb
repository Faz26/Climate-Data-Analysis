```python
Cities=['Bangalore']*8036
df_ban['CITY']=Cities
Cities=['Mumbai']*8036
df_mum['CITY']=Cities
Cities=['Kolkata']*8036
df_kol['CITY']=Cities
Cities=['Bhopal']*8036
df_bhl['CITY']=Cities
Cities=['New_Delhi']*8036
df_del['CITY']=Cities

```

```python
index_ban=[i for i in range(8036)]
index_ban=pd.Series(index_ban)
df_ban['INDEX']=index_ban
```

```python
index_mum=[i for i in range(8036,8036*2)]
index_mum=pd.Series(index_mum)
df_mum['INDEX']=index_mum

index_kol=[i for i in range(8036*2,8036*3)]
index_kol=pd.Series(index_kol)
df_kol['INDEX']=index_kol

index_bhl=[i for i in range(8036*3,8036*4)]
index_bhl=pd.Series(index_bhl)
df_bhl['INDEX']=index_bhl

index_del=[i for i in range(8036*4,8036*5)]
index_del=pd.Series(index_del)
df_del['INDEX']=index_del
```
